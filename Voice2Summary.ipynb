```python
# Install the AssemblyAI library
!pip install assemblyai
import assemblyai as aai

# Replace with your AssemblyAI API token
aai.settings.api_key = "YOUR_ASSEMBLYAI_API_KEY"

# URL of the file to transcribe (replace with your audio file URL)
FILE_URL = "https://github.com/AssemblyAI-Examples/audio-examples/raw/main/20230607_me_canadian_wildfires.mp3"

# Transcribe the audio
transcriber = aai.Transcriber()
transcript = transcriber.transcribe(FILE_URL)

# Get the transcribed text
transcribed_text = transcript.text

# Print the transcribed text
print(transcribed_text)
```

```python
# Install the Transformers library
!pip install transformers
from transformers import pipeline

# Create a summarization pipeline
summarizer = pipeline("summarization")

# Generate a summary of the transcribed text
summary_text = summarizer(transcribed_text, max_length=50, min_length=15, do_sample=False)
```

```python
# Voice to text
print("\nTranscribed Text:")
print(transcribed_text)

# Text to Summary
print("\nSummary:")
print(summary_text[0]['summary_text'])
```
